```python
import json
import pandas as pd
import matplotlib.pyplot as plt
from pylab import rcParams
import numpy as np
rcParams['figure.figsize'] = [10, 5]
```

```python
current = 'solution_2_cqp_3'

with open(current + '.json') as f:
    data = json.load(f)
res_tab = []

for key, value in data.items():
    last_best = value['history'][-1]
    tmp = dict(last_best)
    tmp.pop('type')
    res_tab.append({'algo': key, **tmp})
res = pd.DataFrame(res_tab)
res = res.sort_values(by='qc')
res.index = np.arange(1, res.shape[0]+1)
res[['algo', 'step', 'millisFromStart', 'qcEvals', 'modelEvals', 'qc']].to_csv(current + ".csv")

best_algos = res['algo'].values[0:5]
print(best_algos)
res[['algo', 'step', 'millisFromStart', 'qcEvals', 'modelEvals', 'qc']]
```

```python
for key in best_algos:
    history  = pd.DataFrame(data[key]['history'])
    x = history['millisFromStart'].values
    y = history['qc'].values
    plt.plot(x, y, label=key, linewidth=2)
plt.xlabel('millis')
plt.ylabel('QC')
plt.legend()
plt.savefig(current + '.png')
plt.show()
```

```python
tabu = {
    'avgTime': 0,
    'avgQcEval': 0,
    'avgModelEval': 0
}

metropolis = {
    'avgTime': 0,
    'avgQcEval': 0,
    'avgModelEval': 0
}
for i in range(1, 3):
    for j in range(1, 4):
        current = 'solution_{}_cqp_{}'.format(i, j)
        with open(current + '.json') as f:
            data = json.load(f)
        res_tab = []

        for key, value in data.items():
            last_best = value['history'][-1]
            tmp = dict(last_best)
            tmp.pop('type')
            res_tab.append({'algo': key, **tmp})
        for r in res_tab:
            if r['algo'] == 'TABU_50':
                tabu['avgTime']+=r['millisFromStart']
                tabu['avgQcEval']+=r['qcEvals']
                tabu['avgModelEval']+=r['modelEvals']
            if r['algo'] == 'METRO_50':
                metropolis['avgTime']+=r['millisFromStart']
                metropolis['avgQcEval']+=r['qcEvals']
                metropolis['avgModelEval']+=r['modelEvals']
tabu['avgTime']/=6.0
tabu['avgQcEval']/=6.0
tabu['avgModelEval']/=6.0
tabu['algo'] = 'TABU_50'

metropolis['avgTime']/=6.0
metropolis['avgQcEval']/=6.0
metropolis['avgModelEval']/=6.0
metropolis['algo'] = 'METRO_50'

pd.DataFrame([tabu, metropolis])[['algo', 'avgTime', 'avgQcEval', 'avgModelEval']].to_csv('algo_compare.csv')
```
